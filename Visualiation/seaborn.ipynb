### Load Datasets

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
planets = sns.load_dataset('planets')
```

### Scatterplot

```python
sns.scatterplot(x='tip', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day')
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size')
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette='YlGnBu')
```

### Histograms

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

### Barplot

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

### BoxPlot

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

### StripPlot

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

### JointPlot

```python
sns.jointplot(x='tip', y='total_bill', data=tips)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='YlGnBu')
```

### PairPlot

```python
sns.pairplot(titanic.select_dtypes(include=['number']), hue='pclass')
```

### HeatMap Correlation

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(), annot=True, cmap='YlGnBu')
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(), annot=True, cmap='coolwarm')
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(), annot=True, cmap='icefire')
```
